def generate_content_with_retry(prompt, max_retries=5):
    """
    Envoie une requ√™te √† l'API Gemini en utilisant l'instruction syst√®me
    MSY CLOUD VISION 2030 et impl√©mente le backoff exponentiel.
    """
    if not API_KEY:

    # Construction du payload
    payload = {
        "contents": [{
            "parts": [{"text": prompt}]
        }],
        "systemInstruction": {
            "parts": [{"text": MSY_CLOUD_MASTER_PROMPT}] # Utilisation du Prompt Ma√Ætre 2030

        "tools": [{"google_search": {}}] # Activation de l'outil Google Search pour l'actualisation des donn√©es
    }

    headers = {
        "Content-Type": "application/json"
    }

    for attempt in range(max_retries):
        try:
            # L'API Key est ajout√©e au param√®tre de requ√™te pour l'ex√©cution dans cet environnement
            response = requests.post(f"{API_URL_BASE}?key={API_KEY}", headers=headers, data=json.dumps(payload))
            response.raise_for_status() # L√®ve une exception pour les codes d'erreur HTTP (4xx ou 5xx)

            result = response.json()
            candidate = result.get('candidates', [{}])[0]
            
            # Extraction du texte g√©n√©r√©
            generated_text = candidate.get('content', {}).get('parts', [{}])[0].get('text', '')
            
            # Extraction des sources de mise √† la terre (Grounding Sources)
            sources = []
            grounding_metadata = candidate.get('groundingMetadata', {})
            attributions = grounding_metadata.get('groundingAttributions', [])
            
            for attr in attributions:
                web = attr.get('web', {})
                if web.get('uri') and web.get('title'):
                    sources.append({
                        'uri': web['uri'],
                        'title': web['title']
                    })

            return generated_text, sources

        except requests.exceptions.RequestException as e:

                print(f"√âchec final de la requ√™te apr√®s {max_retries} tentatives. Erreur: {e}")
                return None, []

    return None, []
